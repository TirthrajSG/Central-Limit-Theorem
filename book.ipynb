# Simulating Central Limit Theorem using different distributions

![](imgs/img1.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

#### We will start by creating a function which produces a histogram of the Sum, along with a normal curve whose mean and variance are found from CLT.

```python
def plot(sample):
    n, sample_size = sample.shape

    # Mean and Variance of the given Distribution
    sample_mean = np.mean(sample)
    sample_var = np.mean(np.var(sample, axis=1))
    
    # Data of Sum of independent rv's
    norm_data = np.sum(sample, axis=1)

    # Mean and Variance of the Sum According to Central Limit Theorem
    norm_mean = sample_size * sample_mean
    norm_sd = np.sqrt(sample_var * sample_size)

    fig, ax = plt.subplots()
    
    # Histogram of the Sum data
    ax.hist(norm_data, bins=50, density=True, alpha=0.6, color='blue', edgecolor='black')
    
    # Normal curve from the mean and variance obtained above
    x = np.linspace(min(norm_data), max(norm_data), 1000)
    ax.plot(x, norm.pdf(x, norm_mean, norm_sd), 'r', linewidth=2, label='Normal Curve')
    
    fig.legend()
    return fig, ax
```

##### Lets start with a Exponential Function

```python
num_samples = 100
sample_size = 30

data = np.random.exponential(5, (num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

This doesn't looks like normal, becuase CLT works for large samples.

##### Lets increase Number of Samples to 1000

```python
num_samples = 1000
sample_size = 30

data = np.random.exponential(5, (num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

It has already started looking like Normal Curve.
##### Lets further increase Number of Samples. 

```python
num_samples = 100000
sample_size = 30

data = np.random.exponential(5, (num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

#### Perfect!, Now lets check for different Distributions.

##### Uniform Distribution

```python
num_samples = 100000
sample_size = 30

data = np.random.uniform(0,1, (num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

##### Binomial Distribution

```python
num_samples = 100000
sample_size = 30

data = np.random.binomial(n=11,p=0.6, size=(num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

##### Poisson Diistribution

```python
num_samples = 100000
sample_size = 30

data = np.random.poisson(lam=4, size=(num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

##### Beta Distribution

```python
num_samples = 100000
sample_size = 30

data = np.random.beta(3, 4, (num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

##### Geometric Distribution

```python
num_samples = 100000
sample_size = 30

data = np.random.geometric(p=0.6, size=(num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

##### Gamma Distribution

```python
num_samples = 100000
sample_size = 30

data = np.random.gamma(2, 1, (num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```

##### Lognormal Distribution

```python
num_samples = 100000
sample_size = 30

data = np.random.lognormal(.2, .3, (num_samples, sample_size))
fig, ax = plot(data)
fig.show()
```
